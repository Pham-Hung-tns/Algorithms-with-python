```python
import math
tick = [] #list
ngto = []

# kiem tra so ngto
def Calculation(n):
  for j in range(2,int(math.sqrt(n))+1,1):
    if n % j == 0:
      return False
  return True

# tim kiem cac so ngto tu 2 den 1000
for i in range(2,1000,1):
  if i**2 < 1000:  
    if Calculation(i) == True:
      ngto.append(i)
  else: break

# tinh toan 
for i in range(2,1001):
  if i**2 <= 1000:  
    tick.append(i**2)
  tam = 0
  if not Calculation(i) and i not in tick: 
    for j in range(0,len(ngto),1):
      if i % ngto[j] == 0:
        if i % ngto[j]**2 == 0:
          tam = 1 
        else: 
          tam = 0 
          break
    if tam == 1 : tick.append(i) 
tick.sort()
print("mang nguyen to: ", ngto)
print(tick)
  
```
